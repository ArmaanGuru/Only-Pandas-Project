```python
import pandas as pd
import os
import glob
```

```python
# glob only show that files which is ends with .xlsx name
files=glob.glob('data/*.xlsx')
files
```

```python
# now merge all months in one dataframe
df_list=[pd.read_excel(f) for f in files]
df=pd.concat(df_list,ignore_index=True)
df.head()
```

```python
df.drop_duplicates(inplace=True)
df.head()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df['City']=df['City'].str.strip().str.title()
df['City'].unique()
```

```python
df['Date']=df['Date'].fillna('2025-01-01')
df
```

```python
df.isna().sum()
```

```python
# to check which row with the nan value?
df[df['Unit_Price'].isna()]
# df['Unit-Price']=df['Unit_Price'].fillna(df['Unit_Price'].mean())
```

```python
df['Product']=df['Product'].str.strip().str.title()
df['Category']=df['Category'].str.strip().str.title()
df['Customer']=df['Customer'].str.strip().str.title()
df['Payment_Mode']=df['Payment_Mode'].str.strip().str.title()
df['Quantity']=df['Quantity'].bfill()
df['Total']=df['Total'].bfill()


df
```

```python
df.isna().sum()
```

```python
df['Category']=df['Category'].replace({'Cloths':'Clothes'})
df['Product'] = df['Product'].str.replace("-", "").str.title()
df['Customer']=df['Customer'].replace({'Arman':'Armaan'})
df['City']=df['City'].replace({'Banglore':'Bangalore'})
df
```

```python
df=df.drop_duplicates(subset=['Order_ID'],keep='first')
df.head()
```

```python
df=df.drop(df.columns[-1],axis=1)
df.head()
```

```python
df['Total']=df['Unit_Price']*df['Quantity']
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'],errors='coerce',dayfirst=True,format='mixed')
```

```python
df.head()
```

```python
# Reset Order_ID to be sequential
df['Order_ID'] = range(2004, 2004 + len(df))
df.head()
```

```python
df['Unit_Price'] = df['Unit_Price'].ffill()
df['Total']=df['Total'].ffill()
```

```python
df
```

```python
df['Total']=df['Unit_Price']*df['Quantity']
df
```

```python
total_revenue = df['Total'].sum()
total_quantity = df['Quantity'].sum()
unique_customers = df['Customer'].nunique()
unique_products = df['Product'].nunique()

print("\n=== BASIC BUSINESS METRICS ===")
print(f"Total Revenue: {total_revenue}")
print(f"Total Quantity Sold: {total_quantity}")
print(f"Unique Customers: {unique_customers}")
print(f"Unique Products: {unique_products}")
```

```python
print("\n=== SALES BY CITY ===")

city_sales=df.groupby('City')['Total'].sum().sort_values(ascending=False)
print(city_sales)
```

```python
print('\n===SALES BY CATEGORY===')
category_sales=df.groupby('Category')['Total'].sum().sort_values(ascending=False)
category_sales
```

```python
print('\n===CUSTOMERS EXPENSES & BEST(TOP) CUSTOMER===')
coustomer_expense=df.groupby('Customer')['Total'].sum().sort_values(ascending=False)
coustomer_expense
```

```python
print('\n===BEST SELLING PRODUCT==')
best_sales=df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
best_sales
```

```python
print('\n===REVENUE BY PRODUCT==')
best_sales=df.groupby('Product')['Total'].sum().sort_values(ascending=False)
best_sales
```

```python
print('\n ===SALES BY PAYMENT_MODE===')
payment_mode=df.groupby('Payment_Mode')['Total'].sum().sort_values(ascending=False)
payment_mode
```

```python

df['Month']=df['Date'].dt.month
print('\n===monthly sales analysis===')
month_sales=df.groupby('Month')['Total'].sum()
month_sales
```

```python
print('\n===Daily sales===')
daily_sales=df.groupby('Date')['Total'].sum()
daily_sales
```

```python
pivot=pd.pivot_table(df,values='Total',index='Category',columns='City',aggfunc='sum',fill_value=0)
print('\n===SALES PIVOT TABLE===')
pivot
```
